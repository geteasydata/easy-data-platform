```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 20)
print(f"Pandas: {pd.__version__}, NumPy: {np.__version__}")
```

## 1. Loading Data

```python
# Create sample data
np.random.seed(42)
df = pd.DataFrame({
    'customer_id': range(1, 101),
    'age': np.random.randint(18, 70, 100),
    'income': np.random.normal(50000, 15000, 100).astype(int),
    'spending_score': np.random.randint(1, 101, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100)
})

print(f"Shape: {df.shape}")
df.head()
```

## 2. Data Inspection

```python
# Basic info
print("ðŸ“Š Dataset Info:")
print(df.info())

print("\nðŸ“ˆ Statistics:")
display(df.describe())

print("\nðŸ”¢ Data Types:")
print(df.dtypes)
```

## 3. Data Selection

```python
# Selecting columns
print("Single column:", df['age'].head().tolist())
print("\nMultiple columns:")
display(df[['age', 'income']].head())

# Filtering rows
high_income = df[df['income'] > 60000]
print(f"\nHigh income customers: {len(high_income)}")
```

## 4. Data Aggregation

```python
# Group by
region_stats = df.groupby('region').agg({
    'income': ['mean', 'std'],
    'spending_score': 'mean',
    'customer_id': 'count'
}).round(2)

print("ðŸ“Š Stats by Region:")
display(region_stats)
```

## 5. NumPy Operations

```python
# NumPy array operations
ages = df['age'].values

print("NumPy Stats:")
print(f"  Mean: {np.mean(ages):.1f}")
print(f"  Median: {np.median(ages):.1f}")
print(f"  Std: {np.std(ages):.1f}")
print(f"  Percentiles: {np.percentile(ages, [25, 50, 75])}")
```

## ðŸŽ¯ Key Takeaways
1. `df.head()`, `df.info()`, `df.describe()` for inspection
2. Square brackets for selection and filtering
3. `groupby()` + `agg()` for aggregation
4. NumPy for fast numerical operations
