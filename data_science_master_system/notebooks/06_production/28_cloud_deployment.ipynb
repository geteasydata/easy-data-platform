## 1. Docker Configuration

```python
dockerfile = '''
FROM python:3.10-slim

WORKDIR /app
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

COPY . .

EXPOSE 8000
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
'''

docker_compose = '''
version: "3.8"
services:
  api:
    build: .
    ports:
      - "8000:8000"
    environment:
      - MODEL_PATH=/app/model.joblib
    volumes:
      - ./models:/app/models
  redis:
    image: redis:alpine
    ports:
      - "6379:6379"
'''

print("üê≥ Dockerfile:")
print(dockerfile)
```
